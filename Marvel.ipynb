```python
import networkx as nx
import numpy as np
from networkx.algorithms import bipartite
import matplotlib.pyplot as plt

#comics.txt and characters.txt was generated using a code and out.txt contains edge-list
characters = np.loadtxt('characters.txt',dtype=int)
comics = np.loadtxt('comics.txt',dtype=int)
B = nx.Graph()
B.add_nodes_from(characters,bipartite=0)
B.add_nodes_from(comics,bipartite=1)
h = nx.read_edgelist('out.txt',create_using=nx.Graph(),nodetype=int)
B.add_edges_from(h.edges)
freq = nx.degree_histogram(B)
degrees = range(len(freq)) # this will give numbers from 0 to highest degree.
plt.loglog(degrees[:], freq[:],'g.') #to plot the data in logarithmic scale on both the axes
plt.xlabel('Degree')
plt.ylabel('Frequency')
plt.grid()
plt.savefig("Degree_dis.png")
#The below code may be used for future calculations
#deg = B.degree()
# sum = 0
# for i in B.nodes():
#     if deg[i]>2:
#         sum+=1
# print(sum)
# c = nx.clustering(B)
#print("Clustering coefficient = ",c)
```
